```python
# temps en seconde
T = 2 # secondes
fs = 44100 # Hertz

# t = np.arange(T*fs)/fs
t = np.arange(0, T, 1/fs)

y = np.sin(t)

plt.plot(t,y)
```

```python
freq = 15000
y = np.sin(2*np.pi*freq*t)

plt.plot(t,y)
plt.plot(t,y,"or")
plt.xlim(0,0.005)
```

```python
plt.specgram(y,Fs=fs)
```

```python
plt.figure(figsize=(12,4))

s = np.arange(80)
x1D = np.random.randn(80)
x2D = np.random.randn(80,120)

plt.plot(s, x1D, color = [0,1,1], lw=5, alpha=.15, label="gros")

plt.plot(s, x1D, "--k", label="noir") # pointills noirs
plt.plot(s, x1D, ":r", label="rouge") # ligne rouge
plt.plot(s, x1D, "oc", label="ronds")
plt.plot(s, x1D, "dy", label="diams") #diamand
plt.plot(s, x1D, ".m", label="point") # points magenta

plt.legend(loc="upper left")

plt.grid()
```
